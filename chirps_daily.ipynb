```python
import os
import xarray as xr
import pandas as pd
import glob
```

```python
pwd = os.getcwd()
pwd
```

```python
# file_nc = glob.glob(os.path.join(pwd, 'chirps*1990*.nc'))[0]

bound_ = [108.5, 111.8, -8.22, -6.3]  # ini batas jateng

files_nc = glob.glob(os.path.join(pwd, 'chirps-v2*.nc'))

dss = []
for input_ in files_nc:
    print(input_)
    with xr.open_dataset(input_) as dataset:
        prec = dataset['precip']

    if prec.latitude[0].values < prec.latitude[1].values:
        prec = prec.sel(latitude=slice(bound_[2], bound_[3]))
    else:
        prec = prec.sel(latitude=slice(bound_[3], bound_[2]))
    if prec.longitude[0].values < prec.longitude[1].values:
        prec = prec.sel(longitude=slice(bound_[0], bound_[1]))
    else:
        prec = prec.sel(lon=slice(bound_[1], bound_[2]))
    # cropped_ds = prec.where(mask_lon & mask_lat, drop=True)
    dss.append(prec)
merged = xr.merge(dss)
merged
```

```python
chirps_gabung = os.path.join(pwd, 'chirps_jateng_1990_2005.nc')
merged.to_netcdf(chirps_gabung)
```
